```python
arr = np.load('requiredDataFull.npy')
# Split off the extra column (e.g., touchdown distance)
time_series_data = arr[:, :-1]  # shape: (x, 483)
touchdown_distance = arr[:, -1]  # shape: (x,)

# Reshape and compute mean and std
reshaped = time_series_data.reshape(-1, 23, 21)  # shape: (x, 23, 21)
means = reshaped.mean(axis=2)  # shape: (x, 23)
stds = reshaped.std(axis=2)    # shape: (x, 23)

# Optionally include the touchdown distance as a feature
final_features = np.concatenate([means, stds, touchdown_distance[:, None]], axis=1)  # shape: (x, 47)
```
